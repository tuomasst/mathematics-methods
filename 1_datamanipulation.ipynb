```python
kd = pd.read_csv('kd.csv', na_values = '?')
#kd1 = kd.iloc[320:350,7:25]
kd.head()
```

```python
ckd_true = kd['class']=='ckd'
kd1 = kd[ckd_true]
kd1.describe()
```

```python
cm = kd1.corr()
print (cm)
```

```python
import matplotlib.pyplot as plt
import numpy

names = cm.keys()

fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cm, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = numpy.arange(0,12,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```
